# Credit Scoring

## Contexto

Equifax tiene un modelo de credit score de mercado desarrollado específicamente para Instituaciones Microfinancieras reguladas y no reguladas que predice el comportamiento de pago de una persona natural o jurídica, titular de un crédito para Microempresas.

Diferencias entre una persona jurídica y natural:
   
>1.- Una Persona Natural es una persona humana que ejerce derechos y cumple obligaciones a título personal.
    Al constituir una empresa como Persona Natural, la persona (el dueño de la empresa) asume a título personal todas las obligaciones de la empresa, lo cual implica que asume la responsabilidad y garantiza con todo el patrimonio que posea (los bienes que estén a su nombre), las deudas que pueda contraer la empresa.
    Si, por ejemplo, la empresa quiebra y es obligada a pagar alguna deuda, la persona deberá hacerse responsable por esta a título personal y, en caso de no pagarla, sus bienes personales podrían ser embargados.
    
>2.-Una Persona Jurídica es una empresa que ejerce sus propios derechos y cumple sus propias obligaciones.
    Al constituir una empresa como Persona Jurídica, es la empresa (y no el dueño o los dueños) quien asume todas sus obligaciones, lo cual implica que las deudas que pueda contraer están garantizadas y se limitan solo a los bienes que pueda tener a su nombre (tanto capital como patrimonio).
    Si, por ejemplo, la empresa quiebra y es obligada a pagar alguna deuda, esta se pagará solo con los bienes que pueda tener la empresa a su nombre, sin poder obligar al dueño o a los dueños a tener que hacerse responsable por la deuda con sus bienes personales.

Ordena las cuentas de acuerdo a su nivel de riesgo, permitiendo a cada empresa definir los criterios de admisión de clientes deseados.

Es útil para todo el ciclo de vida del crédito, lo que permite mejorar la consistencia y precisión de las decisiones crediticias. No requiere datos adicionales de la solicitud, sólo nombre y documento de identidad. Puede utilizarse combinado con reglas de decisión ya utilizadas para aprobar o denegar un crédito de Microfinanzas. De esta forma, no se exige un cambio radical de políticas crediticias.

# Microcréditos


El microcrédito es una de las herramientas más solidarias del sistema financiero. Su finalidad es otorgar una oportunidad a aquellos emprendedores que desean crear su negocio, desarrollarlo o mejorar sus condiciones, pero que por distintas circunstancias, **no pueden acceder a un préstamo por los medios tradicionales.**

Estos microcréditos se caracterizan por ser un préstamo de bajos montos e intereses no tan elevados, **con un plazo de tiempo para pagar relativamente corto.**

En el Perú, el sistema formal de banca no otorga créditos a pequeños negocios, es por esta razón que los microcréditos son la mejor opción, brindando el financiamiento deseado a pequeños negocios con pocos requisitos.
